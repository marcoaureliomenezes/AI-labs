```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Numpy Important Methods

- `np.array()`: Cria um array numpy a partir de uma lista.
- `np.arange()`: Cria um array numpy com valores de um intervalo e um passo definidos.
- `np.linspace()`: Cria um array numpy com valores de um intervalo, com um número de elementos definido.
- `np.zeros()`: Cria um array numpy com valores 0.
- `np.ones()`: Cria um array numpy com valores 1.
- `np.eye()`: Cria um array numpy com valores 0, exceto na diagonal principal, que tem valor 1.
- `np.random.rand()`: Cria um array numpy com valores aleatórios entre 0 e 1.
- `np.random.randn()`: Cria um array numpy com valores aleatórios de uma distribuição normal.
- `np.random.randint()`: Cria um array numpy com valores aleatórios inteiros entre um intervalo definido.


```python
arr_1 = np.array([i for i in range(16)])
print(f"\nShape: {arr_1.shape}, Type: {arr_1.dtype}")
print(arr_1)

arr_zeros = np.zeros((4, 4))
print(f"\nShape Matriz Nula: {arr_zeros.shape}, {arr_zeros.dtype}")
print(arr_zeros)

arr_ones = np.ones((4, 4))
print(f"\nShape Matriz : {arr_ones.shape}, {arr_ones.dtype}")
print(arr_ones)

arr_eye = np.eye(4)
print(f"\nShape Matriz Diagonal: {arr_eye.shape}, {arr_eye.dtype}")
print(arr_eye)

arr_rand = np.random.random((4, 4))
print(f"\nShape Matriz Random: {arr_rand.shape}, {arr_rand.dtype}")
print(arr_rand)
```

### Funções trigonométricas e exponenciais

```python
eixo_x = np.linspace(-2*np.pi, 2*np.pi, 100)

sen_fuct = np.sin(eixo_x)
cos_fuct = np.cos(eixo_x)
tg_fuct = np.tan(eixo_x)

fig, eixo = plt.subplots(1, 3, figsize=(15, 3))
_ = [eixo[i].grid(True) for i in range(3)]
# Include the same label on axis x
_ = [eixo[i].set_xlabel("radianos") for i in range(3)]
_ = [eixo[i].set_title(j) for i, j in enumerate(['Seno', 'Cosseno', 'Tangente'])]

fig.suptitle("Funções Trigonométricas", fontsize=16)
sns.lineplot(x=eixo_x, y=sen_fuct, ax=eixo[0])
sns.lineplot(x=eixo_x, y=cos_fuct, ax=eixo[1])
sns.lineplot(x=eixo_x, y=tg_fuct, ax=eixo[2])

plt.tight_layout()
plt.show()
```
