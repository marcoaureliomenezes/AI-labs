# 1- Leitura dos dados

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
%matplotlib inline

df_data = pd.read_csv('datasets/monitoramento_tempo.csv')
df_data['data'] = pd.to_datetime(df_data['data'])
eixo_x = df_data['data']
eixo_y = df_data['temperatura']

df_data.info()
```

# 2 - Gráficos de séries temporais

##### 2.1 - Eixo x (tempo) completo

```python

plt.figure(figsize=(10,5)) # Tamanho da figura
plt.title('Temperatura no tempo 2012-2018') # Título do gráfico
plt.xlabel('Data') # Nome do eixo x
plt.ylabel('Temperatura') # Nome do eixo y
plt.grid(True) # Adiciona o grid
plt.plot(eixo_x, eixo_y)
```

### Adicionando `variaveis` fig e `eixo`

```python
fig = plt.figure(figsize=(10,5)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo
eixo.plot(eixo_x, eixo_y, color = 'red', lw = 1) # Plota o gráfico. Documentação: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

eixo.set_xlim(dt(2014,1,1), dt(2015,1,1))
eixo.set_title('Temperatura no tempo 2014') # Título do gráfico
eixo.set_xlabel('Data') # Nome do eixo x
eixo.set_ylabel('Temperatura') # Nome do eixo y
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15) # Legenda do gráfico
eixo.grid(True) # Adiciona o grid no gráfico
```

```python
fig = plt.figure(figsize=(10,6)) # Tamanho da figura

eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo
eixo2 = fig.add_axes([0.7,0.65,0.3,0.3]) # Adiciona eixo

eixo.grid(True) # Adiciona o grid no gráfico
eixo2.grid(True) # Adiciona o grid no gráfico

eixo.plot(eixo_x, eixo_y, color = 'g', lw='1')
eixo.set_xlim(dt(2014,1,1), dt(2015,1,1))
eixo.set_title('Temperatura em 2014', fontsize=20, pad=15) # Título do gráfico
eixo.set_xlabel('Data') # Nome do eixo x
eixo.set_ylabel('Temperatura') # Nome do eixo y
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15) # Legenda do gráfico

eixo2.plot(eixo_x, eixo_y, color = 'b')
eixo2.set_title('Temperatura no tempo', fontsize=8) # Título do gráfico
eixo2.legend(['Temperatura'], loc = 'best', fontsize=10) # Legenda do gráfico
eixo2.set_xlabel('Data') # Nome do eixo x
eixo2.set_ylabel('Temperatura') # Nome do eixo y
```

```python
fig = plt.figure(figsize=(10,6)) # Tamanho da figura

eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo
eixo2 = fig.add_axes([0.7,0.65,0.3,0.3]) # Adiciona eixo

eixo.grid(True) # Adiciona o grid no gráfico
eixo2.grid(True) # Adiciona o grid no gráfico

eixo.plot(eixo_x, eixo_y, color = 'g', lw='1')
eixo.set_xlim(dt(2014,5,1), dt(2014,6,1))
eixo.set_ylim(270, 330)
eixo.set_title('Temperatura em maio de 2014', fontsize=20, pad=15) # Título do gráfico
eixo.set_xlabel('Data') # Nome do eixo x
eixo.set_ylabel('Temperatura') # Nome do eixo y
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15) # Legenda do gráfico

eixo2.plot(eixo_x, eixo_y, color = 'b')
eixo2.set_xlim(dt(2014,1,1), dt(2015,1,1))
eixo2.set_title('Temperatura de 2014', fontsize=8) # Título do gráfico
eixo2.legend(['Temperatura'], loc = 'best', fontsize=10) # Legenda do gráfico
eixo2.set_xlabel('Data') # Nome do eixo x
eixo2.set_ylabel('Temperatura') # Nome do eixo y
```

```python
fig = plt.figure(figsize=(13,6)) # Tamanho da figura

eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo
eixo2 = fig.add_axes([0.7,0.65,0.3,0.3]) # Adiciona eixo

eixo.grid(True) # Adiciona o grid no gráfico
eixo2.grid(True) # Adiciona o grid no gráfico

eixo.plot(eixo_x, eixo_y, color = 'g', lw='1')
eixo.set_xlim(dt(2014,5,1), dt(2014,6,1))
eixo.set_ylim(270, 330)
eixo.set_title('Temperatura em maio de 2014', fontsize=20, pad=15) # Título do gráfico
eixo.set_xlabel('Data') # Nome do eixo x
eixo.set_ylabel('Temperatura') # Nome do eixo y
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15) # Legenda do gráfico

azul_esquerda = df_data['data'] < dt(2014,5,1)
azul_direita = df_data['data'] > dt(2014,6,1)


eixo2.plot(eixo_x, eixo_y, color = 'g')
eixo2.plot(eixo_x[azul_esquerda], eixo_y[azul_esquerda], color = 'b')
eixo2.plot(eixo_x[azul_direita], eixo_y[azul_direita], color = 'b')
eixo2.set_xlim(dt(2014,1,1), dt(2015,1,1))
eixo2.set_title('Temperatura de 2014', fontsize=8) # Título do gráfico
eixo2.legend(['Temperatura'], loc = 'best', fontsize=10) # Legenda do gráfico
eixo2.set_xlabel('Data') # Nome do eixo x
eixo2.set_ylabel('Temperatura') # Nome do eixo y
```

```python
fig = plt.figure(figsize=(13,6)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo

eixo.plot(eixo_x, eixo_y, color = 'g', lw='1')
eixo.set_title('Temperatura no tempo', fontsize=20, pad=15) # Título do gráfico
eixo.set_xlabel('Data', fontsize=15) # Nome do eixo x
eixo.set_ylabel('Temperatura', fontsize=15) # Nome do eixo y
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15) # Legenda do gráfico
eixo.grid(True) # Adiciona o grid no gráfico

x1 = df_data['data'][df_data['temperatura'].idxmax()]
y1 = max(df_data['temperatura'])
x2 = df_data['data'][df_data['temperatura'].idxmax() - 7000]
y2 = max(df_data['temperatura']) - 10
eixo.annotate('T Máxima', xy=(x1, y1), fontsize=15, xytext=(x2, y2), arrowprops=dict(facecolor='black', shrink=0.05))

x1 = df_data['data'][df_data['temperatura'].idxmin()]
y1 = min(df_data['temperatura'])
x2 = df_data['data'][df_data['temperatura'].idxmin() - 7000]
y2 = min(df_data['temperatura']) + 5
eixo.annotate('T Mínima', xy=(x1, y1), fontsize=15, xytext=(x2, y2), arrowprops=dict(facecolor='black', shrink=0.05))

eixo.axhline(y=max(df_data['temperatura']), color='r', linestyle='--')
eixo.axhline(y=min(df_data['temperatura']), color='r', linestyle='--')
```

```python
temperatura_por_dia_da_semana = df_data.groupby('dia_da_semana').mean('temperatura')
dias_semana = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana.reindex(dias_semana)
```

# Graficos de Barras

```python
fig = plt.figure(figsize=(5,4)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo

indice = [1,2,3,4,5,6,7]
cores = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'brown']
eixo.bar(indice, temperatura_por_dia_da_semana['temperatura'], color=cores, edgecolor='black')


eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10) # Título do gráfico
eixo.set_xlabel('Dia da semana', fontsize=15) # Nome do eixo x
eixo.set_ylabel('Temperatura', fontsize=15) # Nome do eixo y
eixo.set_xticks(indice)
eixo.set_xticklabels(dias_semana)
```

# Gráficos de Pizza

```python
fig = plt.figure(figsize=(5,4)) # Tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # Adiciona eixo

explodir = (0.1, 0, 0, 0, 0, 0.1, 0.1)  

eixo.pie(temperatura_por_dia_da_semana['temperatura'], labels=dias_semana, autopct='%1.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10) # Título do gráfico
```
