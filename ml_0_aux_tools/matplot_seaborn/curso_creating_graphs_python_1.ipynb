```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

PATH = '../../datasets/others/imigrantes_canada.csv'

df_data = pd.read_csv(PATH)
df_data.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
df_brasil = df_data.loc['Brasil', anos]
dict_brasil = {'Ano': df_brasil.index.tolist(), 'Brasil': df_brasil.values.tolist()}
df_brasil = pd.DataFrame(dict_brasil)
df_brasil.head()
```

```python
df_america_do_sul = df_data.query('Região == "América do Sul"')
df_america_do_sul_sorted = df_america_do_sul.sort_values(by='Total', ascending=True, inplace=False)
df_america_do_sul_sorted.head()
```

```python

plt.figure(figsize=(10, 5))
plt.title('Imigrantes Brasileiros no Canadá de 1980 a 2013')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.plot(df_brasil['Ano'], df_brasil['Brasil'], color='green')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # Scala do eixo X
plt.yticks(np.linspace(0, 3000, 7, dtype='int32')) # Scala do eixo Y

plt.show()
```

```python
fig, eixos = plt.subplots(1, 2, figsize=(14, 5))

eixos[0].plot(df_brasil['Ano'], df_brasil['Brasil'], color='green')
eixos[0].set_title('Imigrantes Brasileiros no Canadá\n1980 a 2013')
eixos[0].set_xlabel('Ano')
eixos[0].set_ylabel('Número de imigrantes')
eixos[0].xaxis.set_major_locator(plt.MultipleLocator(5))
eixos[0].grid(True)



plt.show()
```

```python
fig, eixos = plt.subplots(2, 2, figsize=(15, 7))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigrantes no Canadá de 1980 a 2013', fontsize=16)

eixos[0, 0].plot(df_data.loc['Brasil', anos])
eixos[0, 0].set_title('Imigrantes Brasil')

eixos[0, 1].plot(df_data.loc['Colômbia', anos])
eixos[0, 1].set_title('Imigrantes Colômbia')

eixos[1, 0].plot(df_data.loc['Argentina', anos])
eixos[1, 0].set_title('Imigrantes Argentina')

eixos[1, 1].plot(df_data.loc['Peru', anos])
eixos[1, 1].set_title('Imigrantes Peru')

for ax in eixos.flat:
    ax.set(xlabel='Ano', ylabel='Número de imigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid(True)

ymin, ymax = (0, 7000)
for ax in eixos.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

# Gráfico de distribuição

```python
fig, eixos = plt.subplots(1, 2, figsize=(14, 5))

eixos[1].boxplot(df_brasil['Brasil'])
eixos[1].set_title('Boxplot da imigração do Brasil para o canada\n1980 a 2013')
eixos[1].set_xlabel('Brasil')
eixos[1].set_ylabel('Número de imigrantes')
eixos[1].grid(True)

eixos[1].boxplot(df_brasil['Brasil'])
eixos[1].set_title('Boxplot da imigração do Argentina para o canada\n1980 a 2013')
eixos[1].set_xlabel('Brasil')
eixos[1].set_ylabel('Número de imigrantes')
eixos[1].grid(True)
```

## Gráfico de Barras Vertical

```python
fig, eixo = plt.subplots(figsize=(12, 5))

cores = ['royalblue', 'darkorange', 'forestgreen', 'gold', 'firebrick', 'mediumpurple', 'darkcyan', 'saddlebrown', 'gray', 'deeppink']

eixo.bar(df_america_do_sul.index, df_america_do_sul['Total'], color=cores, width=0.5)
eixo.set_title('Imigrantes da América do Sul no Canadá 1980 a 2013', fontsize=18, loc = 'left')
eixo.set_xlabel('')
eixo.set_ylabel('Número de imigrantes', fontsize=14)
eixo.xaxis.set_tick_params(labelsize=10)
eixo.yaxis.set_tick_params(labelsize=10)

plt.show()
```
