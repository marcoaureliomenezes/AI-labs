```python
import pandas as pd
import matplotlib.pyplot as plt


df_data = pd.read_csv('./datasets/imigrantes_canada.csv')

df_data.set_index('Pa√≠s', inplace=True)

anos = list(map(str, range(1980, 2014)))
df_brasil = df_data.loc['Brasil', anos]
dict_brasil = {'Ano': df_brasil.index.tolist(), 'Brasil': df_brasil.values.tolist()}
df_brasil = pd.DataFrame(dict_brasil)
df_brasil.head()
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df_brasil['Ano'], df_brasil['Brasil'], color='green')
plt.title('Imigrantes Brasileiros no Canad√° de 1980 a 2013')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # Scala do eixo X
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000]) # Scala do eixo Y

plt.show()
```

```python
fig, eixo = plt.subplots(figsize=(10, 5))
eixo.xaxis.set_major_locator(plt.MultipleLocator(5))
eixo.set_title('Imigrantes Brasileiros no Canad√°\n1980 a 2013')
eixo.set_xlabel('Ano')
eixo.set_ylabel('N√∫mero de imigrantes')
eixo.plot(df_brasil['Ano'], df_brasil['Brasil'])
plt.show()
```

```python
fig, eixos = plt.subplots(1, 2, figsize=(10, 5))

eixos[0].plot(df_brasil['Ano'], df_brasil['Brasil'], color='green')
eixos[0].set_title('Imigrantes Brasileiros no Canad√°\n1980 a 2013')
eixos[0].set_xlabel('Ano')
eixos[0].set_ylabel('N√∫mero de imigrantes')
eixos[0].xaxis.set_major_locator(plt.MultipleLocator(5))
eixos[0].grid(True)

eixos[1].boxplot(df_brasil['Brasil'])
eixos[1].set_title('Boxplot da imigra√ß√£o do Brasil para o canada\n1980 a 2013')
eixos[1].set_xlabel('Brasil')
eixos[1].set_ylabel('N√∫mero de imigrantes')
eixos[1].grid(True)

plt.show()

```

```python
fig, eixos = plt.subplots(2, 2, figsize=(15, 7))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigrantes no Canad√° de 1980 a 2013', fontsize=16)

eixos[0, 0].plot(df_data.loc['Brasil', anos])
eixos[0, 0].set_title('Imigrantes Brasil')

eixos[0, 1].plot(df_data.loc['Col√¥mbia', anos])
eixos[0, 1].set_title('Imigrantes Col√¥mbia')


eixos[1, 0].plot(df_data.loc['Argentina', anos])
eixos[1, 0].set_title('Imigrantes Argentina')

eixos[1, 1].plot(df_data.loc['Peru', anos])
eixos[1, 1].set_title('Imigrantes Peru')

for ax in eixos.flat:
    ax.set(xlabel='Ano', ylabel='N√∫mero de imigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid(True)

ymin, ymax = (0, 7000)
for ax in eixos.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()

```
