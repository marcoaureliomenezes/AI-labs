```python
import pandas as pd
import matplotlib.pyplot as plt


df_data = pd.read_csv('./datasets/imigrantes_canada.csv')
df_data.set_index('País', inplace=True)

anos = list(map(str, range(1980, 2014)))
df_brasil = df_data.loc['Brasil', anos]
dict_brasil = {'Ano': df_brasil.index.tolist(), 'Brasil': df_brasil.values.tolist()}
df_brasil = pd.DataFrame(dict_brasil)
df_brasil.head()

df_america_do_sul = df_data.query('Região == "América do Sul"')
df_america_do_sul_sorted = df_america_do_sul.sort_values(by='Total', ascending=True, inplace=False)
```

```python
fig, eixo = plt.subplots(figsize=(10, 5))
eixo.xaxis.set_major_locator(plt.MultipleLocator(5))
eixo.set_title('Imigrantes Brasileiros no Canadá\n1980 a 2013')
eixo.set_xlabel('Ano')
eixo.set_ylabel('Número de imigrantes')
eixo.plot(df_brasil['Ano'], df_brasil['Brasil'])
plt.show()
```

```python
fig, eixo = plt.subplots(figsize=(10, 5))
eixo.xaxis.set_major_locator(plt.MultipleLocator(5))
eixo.plot(df_brasil['Ano'], df_brasil['Brasil'], lw=3, color='g')

eixo.set_title('Imigrantes Brasileiros no Canadá 1980 a 2013', fontsize=18, loc = 'left')
eixo.set_xlabel('Ano', fontsize=14)
eixo.set_ylabel('Número de imigrantes', fontsize=14)
eixo.xaxis.set_tick_params(labelsize=10)
eixo.yaxis.set_tick_params(labelsize=10)

eixo.spines['right'].set_visible(False)
eixo.spines['top'].set_visible(False)
plt.grid(ls='--', lw=0.2, c='gray')

fig.savefig('./img/imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
fig, eixo = plt.subplots(figsize=(12, 5))

cores = ['royalblue', 'darkorange', 'forestgreen', 'gold', 'firebrick', 'mediumpurple', 'darkcyan', 'saddlebrown', 'gray', 'deeppink']

eixo.bar(df_america_do_sul.index, df_america_do_sul['Total'], color=cores, width=0.5)
eixo.set_title('Imigrantes da América do Sul no Canadá 1980 a 2013', fontsize=18, loc = 'left')
eixo.set_xlabel('')
eixo.set_ylabel('Número de imigrantes', fontsize=14)
eixo.xaxis.set_tick_params(labelsize=10)
eixo.yaxis.set_tick_params(labelsize=10)

plt.show()
```

```python
df_america_do_sul_sorted = df_america_do_sul.sort_values(by='Total', ascending=True, inplace=False)

fig, eixo = plt.subplots(figsize=(12, 5))

for pais in df_america_do_sul_sorted.index:
    total = df_america_do_sul_sorted.loc[pais, 'Total']
    if pais == 'Brasil':
        eixo.barh(pais, total, color='darkorange')
    else:
        eixo.barh(pais, total, color='gray')

#eixo.barh(df_america_do_sul_sorted.index, df_america_do_sul_sorted['Total'], color=cores)
eixo.set_title('Imigrantes da América do Sul no Canadá 1980 a 2013', fontsize=18, loc = 'left')
eixo.set_xlabel('Número de imigrantes', fontsize=14)
eixo.set_ylabel('')
eixo.xaxis.set_tick_params(labelsize=10)
eixo.yaxis.set_tick_params(labelsize=10)

for i, total in enumerate(df_america_do_sul_sorted['Total']):
    eixo.text(total + 20, i, str(total), fontsize=10, ha='left', va='center', color='black')

eixo.set_frame_on(False)
eixo.get_xaxis().set_visible(False)
eixo.tick_params(axis='both', which='both', length=0)

fig.savefig('./img/imigracao_america_do_sul.png', transparent=False, dpi=300, bbox_inches='tight')


plt.show()
```
