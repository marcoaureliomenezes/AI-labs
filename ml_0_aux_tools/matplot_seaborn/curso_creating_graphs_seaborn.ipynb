```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# supress warnings
import warnings
warnings.filterwarnings('ignore')


df_data = pd.read_csv('./datasets/imigrantes_canada.csv')

df_data.set_index('País', inplace=True)

anos = list(map(str, range(1980, 2014)))
df_brasil = df_data.loc['Brasil', anos]
dict_brasil = {'Ano': df_brasil.index.tolist(), 'Brasil': df_brasil.values.tolist()}
df_brasil = pd.DataFrame(dict_brasil)
df_brasil.head()
```

```python
df_top_10 = df_data.sort_values(['Total',], ascending=False, axis=0).head(10)
df_top_10
```

## 1 - Gráfico de Barras com Seaborn

```python
sns.barplot(data=df_top_10, x=df_top_10.index, y='Total', palette='Paired')
plt.plot()
```

```python
sns.barplot(data=df_top_10, x='Total', y=df_top_10.index)
```

```python
eixo = sns.barplot(data=df_top_10, x='Total', y=df_top_10.index, orient='h')
eixo.set(
    title='Imigrantes Top 10 para o Canada\n1980 - 2013',
    xlabel='Total de imigrantes',
    ylabel='País'
)

plt.plot()
```

```python
fig, eixo = plt.subplots(figsize=(10, 6))
eixo = sns.barplot(data=df_top_10, x='Total', y=df_top_10.index, orient='h')

eixo.set_title('Imigrantes Top 10 para o Canada\n1980 - 2013', fontsize=16)
eixo.set_xlabel('Total de imigrantes', fontsize=14)
eixo.set_ylabel('')

plt.show()
```

```python
def gera_grafico_paleta(palette):
    fig, eixo = plt.subplots(figsize=(10, 6))
    eixo = sns.barplot(data=df_top_10, x='Total', y=df_top_10.index, orient='h', palette=palette)

    eixo.set_title('Imigrantes Top 10 para o Canada\n1980 - 2013', fontsize=16)
    eixo.set_xlabel('Total de imigrantes', fontsize=14)
    eixo.set_ylabel('')

    plt.show()

lista_palhetas = ['Blues_r', 'rocket', 'coolwarm']
gera_grafico_paleta(lista_palhetas[2])
```

```python
palhetas_categoricas = ['Paired', 'Pastel1', 'Pastel2', 'Paired', 'Set1', 'Set2', 'Set3', 'tab10', 'tab20', 'tab20b', 'tab20c']
gera_grafico_paleta(palhetas_categoricas[7])
```

```python
sns.set_theme(style='dark')

gera_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gera_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gera_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gera_grafico_paleta('tab10')
```

```python
fig, eixo = plt.subplots(figsize=(10, 6))
eixo = sns.barplot(data=df_top_10, x='Total', y=df_top_10.index, orient='h', palette='tab10')

eixo.set_title('Imigrantes Top 10 para o Canada\nPeríodo: 1980 - 2013', fontsize=16, loc='left')
eixo.set_xlabel('Total de imigrantes', fontsize=14)
eixo.set_ylabel('')
sns.despine(left=True, bottom=False)

plt.plot()
```
