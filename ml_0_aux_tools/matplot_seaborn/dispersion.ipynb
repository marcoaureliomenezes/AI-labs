```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
from PIL import Image
%matplotlib inline

df_data = pd.read_csv('datasets/iris.csv')
df_data.head()
```

```python
df_data.shape
```

# Gráficos de dispersão

```python
fig = plt.figure(figsize=(10, 7))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': '^'}

for especie in df_data['espécie'].unique():
    df_tmp = df_data[df_data['espécie'] == especie]
    eixo.scatter(df_tmp['comprimento_sépala'], df_tmp['largura_sépala'], color=cores[especie], marker=marcadores[especie])


eixo.set_title('Comprimento da Sépala vs Comprimento da Pétala', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da Sépala', fontsize=15)
eixo.set_ylabel('Largura da Sépala', fontsize=15)
eixo.legend(cores, fontsize=15)
eixo.tick_params(labelsize=15)

```

# Gráfico de distribuição

```python
fig = plt.figure(figsize=(8, 4))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['red', 'green', 'blue', 'orange']

caixas = eixo.boxplot(df_data.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Boxplot da Largura da Pétala', fontsize=25, pad=15)
eixo.set_xticklabels(df_data.drop('espécie', axis=1).columns, fontsize=15)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor, linewidth=2)
    caixa.set(facecolor=cor)

for outlier in caixas['fliers']:
    outlier.set(marker='x', color='black', alpha=0.5)

```

# Gráficos Histogramas

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.grid(True)

eixo.hist(df_data['comprimento_pétala'], bins=20, color='green')
eixo.set_title('Distribuição de comprimento de pétala')
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df_data['comprimento_pétala'].mean(), df_data['comprimento_pétala'].std()

eixo.hist(df_data['comprimento_pétala'], bins=20, color='green')
eixo.set_title('Distribuição de comprimento de pétala')
eixo.grid(True)

eixo.annotate(f'$mu = {round(mu, 2)}$\n$sigma = {round(sigma, 2)}$', xy=(4.5, 20), fontsize=20)
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df_data['comprimento_pétala'].mean(), df_data['comprimento_pétala'].std()

eixo.hist(df_data['comprimento_pétala'], bins=20, color='green')
eixo.set_title('Distribuição de comprimento de pétala')
eixo.grid(True)

eixo.annotate(f'$mu = {round(mu, 2)}$\n$sigma = {round(sigma, 2)}$', xy=(4.3, 20), fontsize=20)

eixo.axvline(mu, color='red', linestyle='--')
eixo.axvline(mu - sigma, color='black', linestyle='--')
eixo.axvline(mu + sigma, color='black', linestyle='--')

eixo.annotate('média', xy=(mu - 0.6, 28), fontsize=20)
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df_data['comprimento_pétala'].mean(), df_data['comprimento_pétala'].std()

eixo.hist(df_data['comprimento_pétala'], bins=20, color='green')
eixo.set_title('Distribuição de comprimento de pétala')
eixo.grid(True)



eixo.axvline(mu, color='red', linestyle='--')
eixo.axvline(df_data['comprimento_pétala'].median(), color='black', linestyle='--')

eixo.annotate(f'$\mu = {round(mu, 2)}$\n$\sigma = {round(sigma, 2)}$', xy=(4.5, 20), fontsize=20)
eixo.annotate('média', xy=(mu - 0.6, 28), fontsize=20)
eixo.annotate('mediana', xy=(df_data['comprimento_pétala'].median() + 0.1, 31), fontsize=20)
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

df_iv = df_data[df_data['espécie'] == 'Iris-versicolor']
mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20)
eixo.set_title('Iris-versicolor')
eixo.grid(True)


eixo.axvline(mu, color='red', linestyle='--')
eixo.axvline(df_iv['comprimento_pétala'].median(), color='black', linestyle='--')


eixo.annotate(f'$\mu = {round(mu, 2)}$\n$\sigma = {round(sigma, 2)}$', xy=(4.5, 20), fontsize=20)
eixo.annotate('média', xy=(mu - 0.5, 5.5), fontsize=20)
eixo.annotate('mediana', xy=(df_iv['comprimento_pétala'].median() - 0.7, 6.5), fontsize=20)

fig.savefig('img/histograma_iv.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

df_is = df_data[df_data['espécie'] == 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins=20)
eixo.set_title('Iris-setosa')
eixo.grid(True)


eixo.axvline(mu, color='red', linestyle='--')
eixo.axvline(df_is['comprimento_pétala'].median(), color='g', linestyle='--')


eixo.annotate(f'$\mu = {round(mu, 2)}$\n$\sigma = {round(sigma, 2)}$', xy=(1.7, 8), fontsize=20)
eixo.annotate('média', xy=(mu - 0.2, 13.5), fontsize=20)
eixo.annotate('mediana', xy=(df_is['comprimento_pétala'].median() + 0.05, 12), fontsize=20, color='g')


fig.savefig('img/histograma_is.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(10, 8))
eixo = fig.add_axes([0, 0, 1, 1])

df_ivc = df_data[df_data['espécie'] == 'Iris-virginica']

mu, sigma = df_ivc['comprimento_pétala'].mean(), df_ivc['comprimento_pétala'].std()

eixo.hist(df_ivc['comprimento_pétala'], bins=20)
eixo.set_title('Iris-Virginica', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)


eixo.axvline(mu, color='red', linestyle='--')
eixo.axvline(df_ivc['comprimento_pétala'].median(), color='g', linestyle='--')


eixo.annotate(f'$\mu = {round(mu, 2)}$\n$\sigma = {round(sigma, 2)}$', xy=(6.3, 7.5), fontsize=20, c='r')
eixo.annotate('média', xy=(mu + 0.05, 8), fontsize=20)
eixo.annotate('mediana', xy=(df_ivc['comprimento_pétala'].median() + 0.05, 7.5), fontsize=20, color='g')


fig.savefig('img/histograma_ivg.png', bbox_inches='tight')
```

```python
largura, altura = Image.open('img/histograma_iv.png').size

nova_imagem = Image.new('RGB', (largura * 3, altura))
intervalo = 0
imagens = ['img/histograma_iv.png', 'img/histograma_is.png', 'img/histograma_ivg.png']
for imagem in imagens:
    img = Image.open(imagem)
    nova_imagem.paste(img, (intervalo, 0))
    intervalo += largura + 5

nova_imagem.save('img/histogramas.png')
```
