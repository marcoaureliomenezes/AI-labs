### Encontrando melhor agrupamento

```python
def plot_perc(st, dados):
    plt.figure(figsize=(20,8))
    g = sns.countplot(x=st, data=dados, palette="Blues_d", orient='h')
    g.set_ylabel("Contagem", fontsize=17)

    sizes=[]
    for p in g.patches:
        height = p.get_height()
        sizes.append(height)
        g.text(p.get_x() + p.get_width()/1.6,
                height + 200,
                '{:1.2f}%'.format(height / 116581*100),
                ha="center", fontsize=12)
        
    g.set_ylim(0, max(sizes) * 1.1)

plot_perc('estados', df_data['estado_cliente'].value_counts())
```
