### Regras de Associação

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df_data = pd.read_csv('../../0_datasets/clustering/dados-padaria.csv', sep=',', header=0)
df_data.head()
```

```python
transacao = []

for item in df_data['TransactionNo'].unique():
    lista = list(set(df_data[df_data['TransactionNo'] == item]['Items']))
    transacao.append(lista)

transacao[:5]
```

```python
te = TransactionEncoder()
te_ary = te.fit(transacao).transform(transacao)
df = pd.DataFrame(te_ary, columns=te.columns_)

itens_frequentes = apriori(df, min_support=0.01, use_colnames=True)
itens_frequentes.sort_values(by='support', ascending=False)
```

```python
regras_apriori = association_rules(itens_frequentes, metric="confidence", min_threshold=0.5)
regras_apriori.sort_values(by='confidence', ascending=False)
```
